```python
from databricks.vector_search.client import VectorSearchClient

vsc = VectorSearchClient()

# Create endpoint (Serverless — pay per query, most cost optimized)
vsc.delete_endpoint(name="hr_policy_vs_endpoint")
vsc.create_endpoint(
    name="hr_policy_vs_endpoint",
    endpoint_type="STANDARD"
)
```

```python
vsc.delete_endpoint(name="hr_policy_vs_endpoint", endpoint_type="STANDARD")
```

```python
# Use Databricks-managed embedding model (no external API cost)
index = vsc.create_delta_sync_index(
    endpoint_name="hr_policy_vs_endpoint",
    source_table_name="workspace.hr.policy_chunks_enriched",
    index_name="workspace.hr.policy_chunks_index",
    pipeline_type="TRIGGERED",  # cost optimized vs CONTINUOUS
    primary_key="doc_id",
    embedding_source_column="chunk_text",
    embedding_model_endpoint_name="databricks-bge-large-en"  # Built-in, no extra cost
)
```
